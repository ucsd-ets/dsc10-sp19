### Example: counts to percent

Create a function that converts an array of counts to percentages of the total
* include docstring

````python
def percents(counts):
    ''' Converts counts to percents out of the total'''
    counts = np.array(counts)  # make sure the container is an array
    total = sum(counts)
    return np.round(counts / total * 100)

percents([1,2,3])
````

# Applying Functions to Tables

## Apply

The `apply` method creates an array by calling a function on every element of the input column(s).
* First argument: Function to apply
* Other argument(s): The input column(s)

```
table_name.apply(function_name, 'column_label')
```

### Use apply to do custom data cleaning
* Very few older people -- replace 3 digit ages with 99

````python
mini_census = Table().with_columns(
    ['Name', ['Ashley', 'Ben', 'Charlie', 'David', 'Ella', 'Frank', 'Greta', 'Henry'], 
     'Age', [27, 68, 106, 51, 102, 27, 115, 2]]
)

mini_census
````

### Use apply to 'clean' dirty data
* Create a function that returns the smallest of `Age` and 99.
* Apply it to all people in census
* Create a new column with cleaned data

````python
def cut_off_at_99(age):
    '''The smaller of x and 100'''
    return min(age, 99)

mini_census.apply(cut_off_at_99, 'Age')
````

````python
mini_census = mini_census.with_column('Age Cleaned', mini_census.apply(cut_off_at_99, 'Age'))
mini_census
````

Define a function `str_len` that takes a string as an input and returns a new string that consists of:
* The given string
* A colon
* "length is"

```
>>> str_len('Data Science rocks!')
```

````python
def str_len(s):
    result = s + ": length is "
    result = result + str(len(s))
    return result

str_len("hats")
````

### Can we predict the height of a child, given the height of their parents?
* read in heights from `galton.csv`
* relabel columns for convenience

````python
#: imports!

import numpy as np
from datascience import *

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
````

````python
galton = (
    Table
    .read_table('galton.csv')
    .relabeled('midparentHeight', 'midparent')
    .relabeled('childHeight', 'child')
)

heights = galton.select('midparent', 'mother', 'father', 'child')
````

````python
heights
````

## Can we predict the height of a child, given the height of their parents?

### scatterplot of child vs. mother/father
* Childrens's height influenced by a combination of the height of the mother and father

````python
heights.select('child', 'mother', 'father').scatter('child')
````

### scatterplot of midparent vs child height
* Galton calculated the variable `midparent` that is a weighted average of the parents' height

````python
heights.scatter('midparent', 'child')
````

### Can we predict the height of a child, given the height of their parents?
* Use the current dataset to inform a prediction of new, unseen parents/children.

````python
heights.scatter('midparent', 'child')
_ = plt.plot([67.5, 67.5], [50, 85], color='red', lw=2)
_ = plt.plot([68.5, 68.5], [50, 85], color='red', lw=2)
_ = plt.scatter(68, 66.24, color='gold', s=40)
````

### Can we predict the height of a child, given the height of their parents?

* Given the midparent height, restrict to nearby examples in the dataset (within 0.5 in).
* Take the average child height within these nearby examples.
* This average is our guess!

````python
def predict_child(mp):
    '''returns a childs predicted height, given the midparent height, mp.'''
    nearby = heights.where('midparent', are.between_or_equal_to(mp - 0.5, mp + 0.5))
    return nearby.column('child').mean()

predict_child(68)
````

### Can we predict the height of a child, given the height of their parents?
* Apply our function to all our examples
* Create a new column called `prediction` and plot the output 

````python
heights = heights.with_column('prediction', heights.apply(predict_child, 'midparent'))
heights
````

````python
heights.select('midparent', 'child', 'prediction').scatter('midparent')
````

# `apply` using Multiple Arguments

### `apply` using Multiple Arguments

The apply method creates an array by calling a function on every element in one or more input columns.
* First argument: Function to apply
* Other argument(s): The input column(s)
* If no columns are supplied, then applies to whole row

```
table_name.apply(function_name, 'column_label1',...,'column_labelN')
```

### Recreate midparent using a function
* Weighted average of father (1.07) and mother (1.0)
* Compare runtime of applying a customer function vs. numpy column operations

````python
def calc_mp(mother, father):
    return (mother + 1.07 * father) / 2
````

````python
% timeit heights.apply(calc_mp, 'mother', 'father')
````

````python
% timeit (heights.column('mother') + heights.column('father') * 1.07)/2
````

### Row objects
* `.row(index)` accesses a row from the table.
* Access an element of a row using `.item(col_name)`

````python
row = heights.row(0)
row
````

````python
row.item('midparent')
````

### Apply a function to row objects
* `.apply(func)` without specifying columns, applies function to entire row.
* Define a function that calculates the prediction errors.

````python
def calc_errors(row):
    return abs(row.item('prediction') - row.item('child'))

heights = heights.with_column(
    'errors',
    heights.apply(calc_errors)
)

heights
````

### Read in NBA dataset

````python
nba = Table.read_table('nba_salaries.csv').relabeled("'15-'16 SALARY", 'SALARY')
nba
````

### Roster sizes
* Count team sizes
* Visualize using a histogram with reasonable bins

````python
nba.group('TEAM').sort('count', descending=True).hist('count', bins=np.arange(9,21))
````

### Team payroll
* Who are the top 5 teams in spending?
* What happens with categorical columns? (clean them up!)

````python
# payroll / drops non-numeric / 
nba.group('TEAM', np.sum)
````

### Which position makes the most on average?
* What about the median? Why the differences?

````python
nba.select('POSITION', 'SALARY').group('POSITION', np.mean).sort('SALARY mean', descending=True)
````

````python
nba.select('POSITION', 'SALARY').group('POSITION', np.median).sort('SALARY median', descending=True)
````

### What is the max salary of each position?
* Apply `max` to grouping (careful of other columns!)
* Can you do this without group?

````python
nba.group('POSITION', max)
````

````python
# with sort
nba.sort('SALARY', descending=True).sort('POSITION', distinct=True)
````

### `list` as an aggregation function 
* Be careful of the size of the table!

````python
nba.where('TEAM', are.containing('Warriors')).group('POSITION', list)
````

### Multiple Groups
* For each position, which team has the most players of that position?
* Pass a list to the first argument of group

````python
nba.group(['POSITION', 'TEAM']).sort('count', descending=True).sort('POSITION', distinct=True)
````

### What are the number of players at each position on *every* team?
* double group with aggregated by counts, followed by single group aggregated by list.
* or, better: `.pivot('POSITION', 'TEAM')`

````python
nba.group(['POSITION', 'TEAM']).group('TEAM', list)
````

````python
nba.pivot('POSITION', 'TEAM')
````

### What is the average salary of each position on each team?
* `pivot` can aggregate values from a third column.

````python

nba.pivot('POSITION', 'TEAM', 'SALARY', np.mean)
````
