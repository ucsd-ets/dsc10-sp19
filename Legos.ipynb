Whether a brick fits on a 3 by 5

```python
def fit35(a, b):
    if a<=3 and b<=5 or a<=5 and b<=3:
        return True
    else:
        return False
    
fit35(4, 2)
```

Use a for loop to create the same array by iterating through each row of the table.

Write a function called size that takes as input two ints and returns a string that is an ordered pair, with the smaller dimension first.

```python
def size(a, b):
    return "("+str(min(a, b))+", "+ str(max(a, b)) +")"

size(6, 4)
```
